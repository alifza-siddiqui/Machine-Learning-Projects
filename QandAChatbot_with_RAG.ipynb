```python
from fpdf import FPDF


def txt_to_pdf(txt_file, pdf_file):
    # Initialize FPDF object
    pdf = FPDF()
    pdf.set_auto_page_break(auto=True, margin=15)
    pdf.add_page()

    # Set font for the PDF (Arial, size 12)
    pdf.set_font("Arial", size=12)

    # Set margins and line height for better readability
    pdf.set_left_margin(10)
    pdf.set_right_margin(10)
    line_height = 10


    with open(txt_file, "r") as file:

        lines = file.readlines()  # Read the content of the .txt file

        # Write each line of the text file into the PDF
        for line in lines:
            # Add line to the PDF with text wrapping
            pdf.multi_cell(0, line_height, line.encode('latin-1', 'replace').decode('latin-1'))

    # Save the PDF
    pdf.output(pdf_file)
    print(f"PDF saved as {pdf_file}")



#convert .txt into .pdf file
txt_to_pdf("business_full_text_wikipedia_data.txt", "business_full_text_wikipedia.pdf")
```

```python
# step 1: convert text into ducment object and read document object
def read_doc(directory):
  loader=PyPDFDirectoryLoader(directory)
  documents=loader.load()
  return documents
```

```python
## step 2: divide doc into chunks
def chunk_data(docs,chunk_size=500,chunk_overlap=40):
  text_splitter=RecursiveCharacterTextSplitter(chunk_size=chunk_size,chunk_overlap=chunk_overlap)
  chunks=text_splitter.split_documents(docs)
  return chunks

documents_chunk=chunk_data(docs=document)
```

```python
## step 3:  vector embedding for this chunk to convert into numerical data
from sentence_transformers import SentenceTransformer

# Initialize the SentenceTransformer model
model = SentenceTransformer('all-MiniLM-L6-v2')
```

```python
# extract plain text for doc chunk which is passed to sentence to for create embedding &
# embedding need plain text not document whic include plain text and plus meta data
text_chunks = [doc.page_content for doc in documents_chunk]
```

```python
# embeddings function
def generate_embeddings_for_chunks(text_chunks):
    embeddings = model.encode(text_chunks)
    return embeddings

# Generate embeddings for each chunk
chunk_embeddings = generate_embeddings_for_chunks(text_chunks)
```

```python
from langchain.chains import LLMChain
from langchain.llms import HuggingFacePipeline
from transformers import pipeline
from langchain.prompts import PromptTemplate
from langchain.schema import Document
from sentence_transformers import SentenceTransformer
```

```python
"""
t5-base
t5-large
facebook/bart-large
google/pegasus-xsum

google/pegasus-large

allenai/longformer-large-4096
"""

# explore all these but google/pegasus-xsum give better result among all !
```

```python
# Function to initialize the QA model pipeline
def initialize_qa_pipeline(model_name="google/pegasus-xsum"):
    return pipeline("text2text-generation", model=model_name)

# Initialize the QA model pipeline by default model used is google/pegasus-xsum
# but you can change this model with any other model
qa_pipeline = initialize_qa_pipeline("google/pegasus-xsum")
```

```python

# Wrap the Hugging Face pipeline with HuggingFacePipeline from Langchain

llm = HuggingFacePipeline(pipeline=qa_pipeline)

# design prompt for effective and accurate response
prompt_template = """
You are an intelligent and trustworthy assistant. Given the provided context, answer the question clearly, concisely, and in grammatically correct sentences.
If the context does not provide sufficient information, use your knowledge to add relevant and accurate details.
However, if you genuinely do not know the answer, state that honestly instead of making up information.
Use the subject-verb-object format in all sentences and ensure proper punctuation.

NOTE: IF YOU ARE NOT 100% SURE OF THE ANSWER, SAY, "I am really sorry but I don't have enough information to answer this question." DO NOT GUESS OR FABRICATE DETAILS UNDER ANY CIRCUMSTANCES.
Make sure the response:

1. Is written in proper English with correct grammar and punctuation.
2. Adheres to the facts and concepts provided in the context.
3. Includes additional knowledge when the context is insufficient (if available).
4. Avoids assumptions or fabricated details.
5. Uses the subject-verb-object format in all sentences and ensures proper punctuation.
6. Access context by given URL if you needed additional information
Example:
Context: {context}

Question: {question}

Answer:
"""

# Creating a PromptTemplate with the defined structure
prompt = PromptTemplate(input_variables=["context", "question"], template=prompt_template)

# Initialize the LLMChain with the prompt and the model -llm
llm_chain = LLMChain(llm=llm, prompt=prompt)

# Function to retrieve relevant documents
# Cosine similary retrive results ie retrive data related to query from vectordb
def retrieve_query(query, k=3):

    # Initialize the model for embeddings (assumed to be Sentence-Transformer or similar)
    model = SentenceTransformer('all-MiniLM-L6-v2')
    query_embedding = model.encode([query])[0]
    query_embedding = query_embedding.tolist()

    # retrieve top k result based on query and store in results
    results = index.query(vector=query_embedding, top_k=k, include_metadata=True)
    return [match['metadata']['text'] for match in results['matches']]



# Function to retrieve and answer a query using LLMChain
def retrieve_and_answer(query):
    # Retrieve relevant documents
    retrieved_contexts = retrieve_query(query)

    # Check if any documents were retrieved
    if not retrieved_contexts:
        return "No relevant documents found."

    # Combine the retrieved documents into one string
    combined_context = " ".join([context.capitalize() + '.' for context in retrieved_contexts])

    # Use the QA pipeline to answer the query
    response = qa_pipeline(combined_context, query)

    response=response[0]['generated_text']

    if len(response.split()) < 5 :  # No context but lengthy response
        return "I am not confident enough to provide an accurate answer."
    if response:
     response = response.capitalize().strip()
     if not response.endswith('.'):
         response += '.'

    return response
```
